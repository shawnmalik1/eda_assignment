```python
# Q2: Airbnb visualizations and summaries (matplotlib only)
import re, numpy as np, pandas as pd, matplotlib.pyplot as plt

# 1
CSV_PATH = "./data/airbnb_hw.csv"
df = pd.read_csv(CSV_PATH)

def norm(s): return re.sub(r'[^a-z0-9]+','', s.lower())
col_map = {norm(c): c for c in df.columns}
room = col_map.get("roomtype") or col_map.get("room_type")
prop = col_map.get("propertytype") or col_map.get("property_type")
price = col_map.get("price")
beds  = col_map.get("beds") or col_map.get("bed") or col_map.get("bedrooms")

# 2) Dimensions, variables, head
print("Shape:", df.shape)
print("Columns:", list(df.columns))
display(df.head())

# 3) Crosstab Room Type x Property Type
ct_counts = pd.crosstab(df[room], df[prop], dropna=False)
ct_rowpct = pd.crosstab(df[room], df[prop], normalize="index", dropna=False)*100
display(ct_counts)
display(ct_rowpct.round(2))

# 4) Price plots + describe; then log-transform and repeat
df[price] = pd.to_numeric(df[price], errors='coerce')
p = df[price].dropna()
display(p.describe(percentiles=[.25,.5,.75]).to_frame("Price"))
plt.figure(); plt.hist(p, bins=50); plt.title("Histogram of Price"); plt.xlabel("Price"); plt.ylabel("Freq"); plt.tight_layout(); plt.show()
try:
    if len(p)>=2:
        ax = p.plot(kind="kde", title="Kernel Density of Price"); ax.set_xlabel("Price"); plt.tight_layout(); plt.show()
except: 
    plt.figure(); plt.hist(p, bins=50, density=True); plt.title("Density proxy (Price)"); plt.tight_layout(); plt.show()
plt.figure(); plt.boxplot(p.values, vert=True, showfliers=True); plt.title("Box plot of Price"); plt.ylabel("Price"); plt.tight_layout(); plt.show()

df["price_log"] = np.where(df[price]>0, np.log(df[price]), np.nan)
pl = df["price_log"].dropna()
display(pl.describe(percentiles=[.25,.5,.75]).to_frame("log(Price)"))
plt.figure(); plt.hist(pl, bins=50); plt.title("Histogram of log(Price)"); plt.xlabel("log(Price)"); plt.ylabel("Freq"); plt.tight_layout(); plt.show()
try:
    if len(pl)>=2:
        ax = pl.plot(kind="kde", title="Kernel Density of log(Price)"); ax.set_xlabel("log(Price)"); plt.tight_layout(); plt.show()
except:
    plt.figure(); plt.hist(pl, bins=50, density=True); plt.title("Density proxy (log Price)"); plt.tight_layout(); plt.show()
plt.figure(); plt.boxplot(pl.values, vert=True, showfliers=True); plt.title("Box plot of log(Price)"); plt.ylabel("log(Price)"); plt.tight_layout(); plt.show()

# 5) Scatter price_log vs Beds; groupby Beds summary
df[beds] = pd.to_numeric(df[beds], errors='coerce')
mask = df["price_log"].notna() & df[beds].notna()
plt.figure(); plt.scatter(df.loc[mask, beds], df.loc[mask, "price_log"], alpha=0.6)
plt.title("Scatter: log(Price) vs Beds"); plt.xlabel("Beds"); plt.ylabel("log(Price)"); plt.tight_layout(); plt.show()
beds_summary = (df.dropna(subset=[beds, price]).groupby(beds)[price]
                .agg(['count','mean','median','std','min','max']).sort_index())
display(beds_summary)

# 6) Colored scatter by Room & Property; group summary
df["_combo"] = df[room].astype(str).str.strip() + " | " + df[prop].astype(str).str.strip()
top = set(df["_combo"].value_counts().head(10).index)
plt.figure()
for combo, sub in df[mask].groupby("_combo"):
    if combo in top:
        plt.scatter(sub[beds], sub["price_log"], label=combo, alpha=0.6)
    else:
        plt.scatter(sub[beds], sub["price_log"], alpha=0.25)
plt.title("Scatter: log(Price) vs Beds (colored by Room & Property)"); plt.xlabel("Beds"); plt.ylabel("log(Price)")
plt.legend(loc="best", fontsize="small"); plt.tight_layout(); plt.show()

rp_summary = (df.dropna(subset=[room, prop, price])
              .groupby([room, prop])[price]
              .agg(['count','mean','median','std','min','max'])
              .sort_values(by='mean', ascending=False))
display(rp_summary)

# Quick highlights
print("\nHighest mean price:", rp_summary['mean'].idxmax(), float(rp_summary['mean'].max()))
print("Highest std (most variable):", rp_summary['std'].idxmax(), float(rp_summary['std'].max()))
print("\nTip: Means are sensitive to outliers and right-skew; medians are more robust, especially for price.")
```

  3. Apartments dominate the dataset, appearing both as entire home/apartment and as private rooms. Private rooms are much more common than entire rentals for certain property types, such as townhouses or shared housing. Entire home/apartments dominate for property types like condominiums or apartments. **Q3.** This question looks at a time series of the number of active oil drilling rigs in the United States over time. The data comes from the Energy Information Agency.

1. Load `./data/drilling_rigs.csv` and examine the data. How many observations? How many variables? Are numeric variables correctly read in by Pandas, or will some variables have to be typecast/coerced? Explain clearly how these data need to be cleaned.
2. To convert the `Month` variable to an ordered datetime variable, use `df['time'] = pd.to_datetime(df['Month'], format='mixed')`.
3. Let's look at `Active Well Service Rig Count (Number of Rigs)`, which is the total number of rigs over time. Make a line plot of this time series. Describe what you see.
4. Instead of levels, we want to look at change over time. Compute the first difference of  `Active Well Service Rig Count (Number of Rigs)` and plot it over time. Describe what you see.
5. The first two columns are the number of onshore and offshore rigs, respectively. Melt these columns and plot the resulting series.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1) Load data
df = pd.read_csv("./data/drilling_rigs.csv")

# 2) Parse date
df['time'] = pd.to_datetime(df['Month'], format='mixed')

# Coerce numeric columns (treat 'Not Available' as NaN)
for c in df.columns:
    if c not in ('Month', 'time'):
        df[c] = pd.to_numeric(df[c], errors='coerce')

# 3) Level plot: Active Well Service Rig Count
active_col = "Active Well Service Rig Count (Number of Rigs)"
plt.figure()
plt.plot(df['time'], df[active_col])
plt.title("Active Well Service Rig Count (levels)")
plt.xlabel("Time"); plt.ylabel("Number of rigs"); plt.tight_layout(); plt.show()

# 4) First difference
df['active_diff'] = df[active_col].diff()
plt.figure()
plt.plot(df['time'], df['active_diff'])
plt.title("Change in Active Well Service Rig Count (first difference)")
plt.xlabel("Time"); plt.ylabel(" rigs (month-to-month)")
plt.axhline(0, linewidth=1); plt.tight_layout(); plt.show()

# 5) Melt first two (onshore/offshore) and plot
onshore_col  = df.columns[1]
offshore_col = df.columns[2]
melt_df = df[['time', onshore_col, offshore_col]].melt(
    id_vars='time',
    value_vars=[onshore_col, offshore_col],
    var_name='Rig type', value_name='Rigs'
)
plt.figure()
for name, sub in melt_df.groupby('Rig type'):
    plt.plot(sub['time'], sub['Rigs'], label=name)
plt.title("Onshore vs Offshore rigs over time")
plt.xlabel("Time"); plt.ylabel("Number of rigs"); plt.legend(loc='best')
plt.tight_layout(); plt.show()
```
